```python
import pandas as pd
import matplotlib.pyplot as plt

with open('/content/episode_data.txt', 'r') as file:
    data_lines = file.readlines()


columns = [
    'Ep_number', 'Ep_Reward', 'Ep_tot_steps', 'T_overlap', 'T_nonbonded',
    'T_tot_all_min', 'T_similar_min', 'T_lower_ene_min', 'T_higher_ene_min',
    'T_unique_min', 'GM_ene', 'atom_select', 'atom_shift'
]


data = []

for line in data_lines:
    values = line.strip().split(', ')
    data_dict = dict(zip(columns, values))
    data.append(data_dict)

df = pd.DataFrame(data)

print(df)
```
